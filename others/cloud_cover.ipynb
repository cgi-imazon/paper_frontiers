```python
import pandas as pd
from glob import glob


pd.set_option('float_format', '{:f}'.format)
```

## Config Data

```python


PATH_CLOUD_COVER = 'data/cloud'
```

## Input Data

```python
df_area = pd.concat([pd.read_csv(x) for x in glob(f'{PATH_CLOUD_COVER}/*')])

df_area = df_area[['year', 'area_ha']]
df_area.columns = ['year', 'area']

df_area['area'] = df_area['area'].div(100)
df_area['area_p'] = df_area['area'].div(4200600)
```
